```python
import nltk  #for stopwords

import numpy as np 

from collections import Counter #to count frequency
from nltk.corpus import stopwords #for stopwords
stopwords=set(stopwords.words('english')) #all unique stopwords

import pandas as pd
from sklearn.datasets import fetch_20newsgroups #fetching the present dataset
from sklearn.model_selection import train_test_split # for spliting the data



```

```python
# BY USING THE INBUILT DATASET AND THE INBUILT FUNCTION AND VECTORIZER (method 1)

news=fetch_20newsgroups()
c=news.data
q=news.target
z=news.target_names #for getting the names of the target
z



```

```python
f_train,f_test,r_train,r_test=train_test_split(c,q,random_state=1) #spilting the test and train data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

#Train and evaluate the model

vect = CountVectorizer(stopwords).fit(f_train) 
#counting the number of each word that appers in a document.
#fit is used to learn a vocabulary from one or more documents
#stopwords function is there to ignore the stop words

f_train_vectorized = vect.transform(f_train)
# transform function is used to on one or more documents as needed to encode each as a vector.

# the calling of inbuilt functions multinomial text classification
clfrNB = MultinomialNB(alpha = 0.1)
clfrNB.fit(f_train_vectorized, r_train)
r_pred = clfrNB.predict(vect.transform(f_test))

#to print the classification report and the confusion matrix
print(classification_report(r_test,r_pred))
inbuilt_text_matrix=confusion_matrix(r_test,r_pred)
print(inbuilt_text_matrix)

```

```python

#to print the accuracy score
from sklearn.metrics import  accuracy_score 
print(accuracy_score(r_test, r_pred))
```

```python
# Fit function-It will make dictionary ,dictionary with top keys as classes and having nested dictionary with keys as 
#feature nos and value as frequency of that feature in respective classes.
                
def fit(X_train, Y_train):
    result = {}
    class_values = Y_train
    for current_class in class_values:
        result[current_class] = {}
        result["total_data"] = 19997
        num_features = X_train.shape[1]
        result[current_class]["total_count"] = data[current_class-1].sum()
        for j in range(num_features ):
            result[current_class][j] = {}
            result[current_class][j] = data[current_class-1][j]  
                                                                 
                                
                
    return result 
```

```python
# It is probability function which will calculate probability

def probability(dictionary, x, current_class):
    output=np.log(1/20)
    num_features = len(dictionary[current_class].keys()) - 1;
    for j in range(num_features ):
        if x[j]>0 :
            count_current_class_with_value_xj = dictionary[current_class][j] + 1
            count_current_class = dictionary[current_class]["total_count"] + num_features
            current_xj_probablity = np.log(count_current_class_with_value_xj)-np.log( count_current_class)
            output = output + current_xj_probablity
    return output
```

```python
# For finding best classes

def predictSinglePoint(dictionary, x):
    class_values = dictionary.keys()
    best_p = -1e100
    best_class = -1
    first_run = True
    for current_class in class_values:
        if (current_class == "total_data"):
            continue
        p_current_class = probability(dictionary, x, current_class)
        if ( first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```

```python
#Predict funcyion for predictions

def predict(dictionary, X_test):
    Y_predalgo = []
    for x in X_test:
        x_class = predictSinglePoint(dictionary, x)
        Y_predalgo.append(x_class)
    return Y_predalgo
```

```python
dictionary = fit(X_train,Y_train)
```

```python
# FOR FINDING THE ACCURACY SCORE OF THE SELF IMPLEMENTATED TEXT CLASSIFICATION

Y_predalgoo = predict(dictionary,X_test)
accuracy_score(Y_predalgoo,Y_test)
```
